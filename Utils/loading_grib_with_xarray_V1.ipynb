The goal of this notebook is to demonstrate how to load a grib file using xarray and then convert it into a pandas data frame

```python
import numpy as np
import pandas as pd
import xarray as xr
```

```python
# Use xarray to load the file ../Data/3b47aa95fa8771c6da86f36140fda074/data.grib

# Specify the path to the GRIB file
grib_file_path = '../Data/3b47aa95fa8771c6da86f36140fda074/data.grib'

# Load the GRIB file using xarray with the cfgrib engine
dataset = xr.open_dataset(grib_file_path, engine='cfgrib')
```

```python
#Convert dataset into a pandas dataframe
df = dataset.to_dataframe()
```
