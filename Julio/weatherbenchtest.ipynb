```python
import xarray as xr
import pandas as pd
```

You can find all the `weatherbench` data [here](https://console.cloud.google.com/storage/browser/weatherbench2/datasets/era5?pageState=(%22StorageObjectListTable%22:(%22f%22:%22%255B%255D%22))).

```python
wb2data = xr.open_zarr('gs://weatherbench2/datasets/era5/1959-2023_01_10-wb13-6h-1440x721.zarr')
filtered_data = wb2data.sel(time=slice('2023-01-01', '2023-01-31'), latitude=slice(40,30), longitude=slice(80,90))
del wb2data
filtered_data
```

```python
features = ['leaf_area_index_high_vegetation','leaf_area_index_low_vegetation','temperature','surface_pressure','vertical_velocity','u_component_of_wind','v_component_of_wind','total_precipitation_6hr']

#features_derived = ['leaf_area_index_high_vegetation','leaf_area_index_low_vegetation','wind_speed','temperature','surface_pressure','vorticity','vertical_velocity','u_component_of_wind','v_component_of_wind','total_precipitation_6hr']

filtered2_data = filtered_data[features]
```
